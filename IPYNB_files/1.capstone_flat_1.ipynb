link column is not necessary for me . i droped it

in society column some society name are like 'Signature Global Solera3.7 ★' i.e.with rating. so i have to fix it

```python
df['society']=df['society'].apply(lambda name: re.sub(r'\d+\.\d+\s★', '', str(name)).strip()).str.lower()
```

In society column two values are as 'not applicable' , '.' , 'nan' and 'nul'  . So I replaced them by the correct name manually.  

```python
df.info()
```

```python
df.loc[df[df.society=='not applicable'].index, 'society'] = 'rk homes'
df.loc[df[df.society=='.'].index, 'society'] = 'vashist properties'
df.loc[df[df.society=='nan'].index, 'society'] = 'raheja revanta'
df.loc[df[df.society=='nul'].index, 'society'] = 'vally view estate'
```

```python
df.society.value_counts().shape
```

in price column some value like 'Price on Request' so i am going to drop them ultimately they are not going to help me.

```python
df.drop(df[df.price == 'Price on Request'].index, inplace = True)
```

in price column some value is null, and i can not fill them randomly it may produce baisness in the time of prediction. So i droped it

```python
df.drop(df[df.price.isnull()].index,inplace = True)
```

In price column some price are in crore and some of them in lac. So i convert them in crore to maintain the same unit.

```python
df['price']=df['price'].str.split(" ").apply(lambda x:float(x[0]) if x[1]=='Crore' else (float(x[0])*100000)/10000000)
```

I renamed the price column as 'price_in_cr'. It identifies that the values present in the column are in crore

```python
df.rename(columns={'price':'price_in_cr'}, inplace=True)
```

```python
df.sample(5)
```

Area column have this "₹ 8,450/sq.ft." kind of value. So i extracted only 8450 to easy to make the prediction

```python
df['area']=df['area'].apply(lambda text:float(''.join(filter(str.isdigit, text))))
```

### ADDITIONAL ROOM

### FLOOR NUMBER

floorNum column contain two null values. So we have to fill them.

```python
df.floorNum.isnull().sum()
```

The floor num is given in the address column for both the rows 181 and 2764

```python
df.loc[181, 'floorNum'] = '6th floor'
df.loc[2764, 'floorNum'] = '2th floor'
```

I want to extract only the floor number. There have somme abnormal value like 'Ground', 'Lower' and 'Basement'.

```python
df['floorNum'].str.split(" ")
```

I replaced Groud as 0, Lower as 0 and Basement as -1 and extract only the floor number.

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```
